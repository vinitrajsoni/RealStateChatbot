```python
df = pd.read_csv('cleaned_file.csv')
df.slug.head(10)
```

```python
def extract_city_locality(slug):
    parts = slug.split("-")
    if len(parts) < 3:
        return "Unknown", "Unknown"
    
    city = parts[-2]  # second last word â†’ city
    locality = " ".join(parts[2:-2]) if len(parts) > 4 else "Unknown"
    return city.capitalize(), locality.replace("-", " ").title()
```

```python
df[["City", "Locality"]] = df["slug"].apply(
    lambda x: pd.Series(extract_city_locality(str(x)))
)
```

```python
df["City + Locality"] = df["City"] + " - " + df["Locality"]
```

```python
df = df.drop(['slug'], axis = 1)
df.head()
```

```python
import pandas as pd


# âœ… Step 1: Define the formatting function
def format_price(price):
    try:
        price = float(price)
    except:
        return "Unknown"
    
    if price >= 1e7:  # 1 crore = 10,000,000
        return f"â‚¹{round(price / 1e7, 2)} Cr"
    elif price >= 1e5:  # 1 lakh = 100,000
        return f"â‚¹{round(price / 1e5, 2)} L"
    else:
        return f"â‚¹{int(price)}"

# âœ… Step 2: Apply formatting
df_new["price"] = df_new["price"].apply(format_price)



print("âœ… Price column formatted and saved as 'Price (formatted as â‚¹X Cr / â‚¹X L)'")
```

```python
df_new = df_new.drop(['Price (formatted as â‚¹X Cr / â‚¹X L)'], axis=1)
```

```python
import pandas as pd
import re


# âœ… Clean the propertyImages column
def clean_image_link(value):
    if pd.isna(value):
        return "Unknown"
    # Remove brackets and quotes using regex
    cleaned = re.sub(r'[\[\]\']""', '', str(value)).strip()
    return cleaned

df_new["propertyImages"] = df_new["propertyImages"].apply(clean_image_link)

# âœ… Save cleaned CSV
df_new.to_csv("cleaned_property_images.csv", index=False, encoding="utf-8-sig")

print("âœ… propertyImages column cleaned â€” brackets removed and file saved as 'cleaned_property_images.csv'")
```
